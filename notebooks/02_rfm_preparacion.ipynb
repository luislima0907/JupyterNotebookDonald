# 02 - Preparacion de Datos RFM
Limpieza, validacion, calculo de puntajes (1-5) y estandarizacion de las
tres dimensiones para el clustering.

## Importar dependencias

```python
import pandas as pd
import numpy as np
import sys
import os

sys.path.append("../src")
from preparacion import (limpiar_datos, validar_rfm,
                         calcular_scores_rfm, preparar_para_clustering)
from conexion import guardar_csv
```

## Cargar los datos extraidos

```python
df = pd.read_csv("../data/clientes_rfm.csv")
print(f"Datos cargados. Filas iniciales: {len(df)}")
df.head()
```

## Limpiar datos

```python
df = limpiar_datos(df)
print(f"Filas despues de limpieza: {len(df)}")
```

## Validar valores RFM

```python
df = validar_rfm(df)
```

## Calcular puntajes de Proximidad, Frecuencia y ValorMonetario (escala 1-5)

```python
df = calcular_scores_rfm(df)
print(df[['CodigoCliente', 'Proximidad', 'Frecuencia', 'ValorMonetario',
          'Puntaje_Proximidad', 'Puntaje_Frecuencia', 'Puntaje_ValorMonetario',
          'Puntaje_RFM']].head(10))
```

## Estandarizar los datos para clustering

```python
X_scaled, scaler = preparar_para_clustering(df)
print("Datos estandarizados (primeras filas):")
print(X_scaled[:10])
print(f"\nMedia (~0): {X_scaled.mean(axis=0)}")
print(f"Desviacion estandar (~1): {X_scaled.std(axis=0)}")
```

## Guardar el dataset preparado

```python
guardar_csv(df, "../data/clientes_rfm_preparado.csv")
np.save("../data/X_rfm_scaled.npy", X_scaled)
print("Archivos clientes_rfm_preparado.csv y X_rfm_scaled.npy guardados")
```
