```python
from src.conexion import conectar_bd, ejecutar_query
from dotenv import load_dotenv
import os

load_dotenv("../.env")

SERVER = os.getenv("SERVER")
DATABASE = os.getenv("DATABASE")
USER = os.getenv("USER")
PASSWORD = os.getenv("PASSWORD")

conexion = conectar_bd(
    SERVER,
    DATABASE,
    USER,
    PASSWORD
)

df_rfm = ejecutar_query(conexion, query)

print("\nValores nulos:")
print(df_rfm.isnull().sum())

print("\nTipos de datos:")
print(df_rfm.dtypes)
print("\nEstadisticas basicas:")
print(df_rfm.describe())

print("\nForma del dataset:")
print(df_rfm.shape)
print("\nColumnas:")
print(df_rfm.columns.tolist())
print("\nPrimeras filas:")
df_rfm.head()
```
