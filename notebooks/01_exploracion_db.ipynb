# 01 - Extraccion de Datos RFM (Taller Donald)
Conexion a SQL Server, ejecucion de la consulta RFM y guardado del
dataset crudo con las tres dimensiones: **Proximidad**, **Frecuencia** y **ValorMonetario**.

## Configuracion del entorno

```python
import os
import sys

sys.path.append(os.path.abspath(".."))

import pandas as pd
import numpy as np
import pyodbc
from dotenv import load_dotenv

from src.conexion import (conectar_bd, ejecutar_query,
                          cargar_query_desde_archivo, guardar_csv)

print("Drivers ODBC disponibles:", pyodbc.drivers())
```

## Conexion a la base de datos
Las credenciales se leen desde el archivo `.env` en la raiz del proyecto.

```python
ruta_env = os.path.abspath("../.env")
load_dotenv(ruta_env, override=True)

SERVER = os.getenv("SERVER")
DATABASE = os.getenv("DATABASE")
USER = os.getenv("USER")
PASSWORD = os.getenv("PASSWORD")

print("SERVER:", SERVER)
print("DATABASE:", DATABASE)

conexion = conectar_bd(SERVER, DATABASE, USER, PASSWORD)
```

## Cargar la consulta RFM

```python
query = cargar_query_desde_archivo("../sql/query_rfm.sql")
print(query)
```

## Ejecutar la consulta y explorar los datos

```python
df_rfm = ejecutar_query(conexion, query)

print("\nForma del dataset:", df_rfm.shape)
print("\nColumnas:", df_rfm.columns.tolist())
print("\nValores nulos:")
print(df_rfm.isnull().sum())
print("\nEstadisticas RFM:")
print(df_rfm[['Proximidad', 'Frecuencia', 'ValorMonetario']].describe())
df_rfm.head(10)
```

## Guardar el dataset crudo y cerrar la conexion

```python
guardar_csv(df_rfm, "../data/clientes_rfm.csv")

conexion.close()
print("Conexion cerrada")
```
